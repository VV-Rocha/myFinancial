```julia
using MarketData
MD = MarketData

using Plots

using Statistics

include("./Data Processing/data_processing.jl")

using .DataDownload
DD = DataDownload

include("./Tsallis/Tsallis.jl")
using .Tsallis
using .qGaussian
```

# Download data

```julia
p2 = Dates.now()
p1 = Dates.now()-Year(3)
interval="1d"
    
symbols = ["MSFT"]#, "TSLA"]
stocks = DD.set_download(symbols, p1=p1, p2=p2, interval=interval)
```

# Growth Ratio

```julia
function growth(data)
    return data[:, 2] ./ data[:, 1]
end

## Plot
plot()
for symbol in symbols
    G = growth(stocks[symbol])
    plot!(stocks[symbol][:, 3], G, linestyle=:solid, label=symbol)
    print(symbol, "=", mean(G), "\n")
end
title!("Growth Rate")
xlabel!("time (s)")
ylabel!("G", rotation=0)
```

```julia
## Plot
plot()
for symbol in symbols
    G = growth(stocks[symbol])
    stephist!(G, bins=32, normalize=:pdf, label=symbol)
    plot!(LinRange(0.85, 1.15, 100), qGaussian.distribution.(LinRange(0.85, 1.15, 100), mean(G), var(G), 1.))
    print(symbol, "=", mean(G), "\n")
end
title!("Growth Rate")
xlabel!("G")
#ylabel!("G", rotation=0)
```

```julia
using StatsBase

r = stocks["MSFT"][:, 1] .* 0.

lags = collect(0:2:size(stocks["MSFT"][:,1])[1]-1)

plot(lags, StatsBase.crosscov(growth(stocks["MSFT"]), growth(stocks["TSLA"]), lags; demean=true))
```

```julia
plot()
for symbol in symbols
    plot!(lags, StatsBase.autocov(growth(stocks[symbol]), lags, demean=true), label=symbol)
end
title!("AutoCov")
xlabel!("lag")
ylabel!("AutoCorrelation", rotation=0)
```
